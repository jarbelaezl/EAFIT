```python
# Importamos las librerías que vamos a utilizar
import pandas as pd
import matplotlib.pyplot as plt

# En Python, una lista se define con corchetes [], y un diccionario con llaves {}
# Aquí usamos un diccionario para guardar los datos del conteo de polen por familia
data = {
    'Familia': ['Urticaceae cecropia', 'Rhamnaceae', 'Dilleniaceae', 'Melastomataceae',
                'Cannabaceae Trema', 'Myrtaceae', 'Polypodiaceae', 'Malvaceae  Heliocarpus',
                'Fabaceae macrolobium', 'Poaceae', 'Malvaceae Bombacoideae', 'Loranthaceae oryctanthus'],
    'Conteo': [845, 120, 56, 39, 12, 21, 10, 7, 8, 6, 5, 4]
}

# Convertimos el diccionario en un DataFrame (tabla de datos) usando pandas
df = pd.DataFrame(data)

# Calculamos el porcentaje que representa cada tipo de polen respecto al total
df['Porcentaje'] = df['Conteo'] / df['Conteo'].sum() * 100

# Ordenamos el DataFrame de mayor a menor según el porcentaje
df.sort_values(by='Porcentaje', ascending=False)

# Mostramos la tabla ordenada
print(df)
```

```python
plt.figure(figsize=(10,6))
plt.barh(df['Familia'], df['Porcentaje'], color='turquoise')
plt.xlabel('% de polen')
plt.title('Distribución porcentual del polen por familia')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
